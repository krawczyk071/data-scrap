```python
import pandas as pd
df = pd.read_csv('../data/output20230720-153319.csv',delimiter=';')

def intround(val:str):
    return int(round(float(val.replace(',','.'))))

df['sqm'] = df['sqm'].str.replace(' m²','').astype(float)
df['perm'] = df['perm'].str.replace(' zł/m²','').str.replace(' ','').fillna('-1').apply(intround)
df['price'] = df['price'].str.replace(' zł','').str.replace('Zapytaj o cenę','-1').str.replace(' ','').fillna('-1').apply(intround)
df['rooms'] = df['rooms'].str.replace(r' pokoje| pokoi| pokój','',regex=True).apply(intround)
df['id'] = df['link'].str[-7:]
```
